# Employee Salary Calculator

```python


def calculate_hra(basic_salary):
    return basic_salary * 0.20

def calculate_da(basic_salary):
    return basic_salary * 0.10

def calculate_net_salary(basic_salary):
    hra = calculate_hra(basic_salary)
    da = calculate_da(basic_salary)
    return basic_salary + hra + da

name = input("Enter Employee Name: ")
basic_salary = float(input("Enter Basic Salary: "))

hra = calculate_hra(basic_salary)
da = calculate_da(basic_salary)
net_salary = calculate_net_salary(basic_salary)

print("\n------ Salary Details ------")
print("Employee Name :", name)
print("Basic Salary  :", basic_salary)
print("HRA (20%)     :", hra)
print("DA (10%)      :", da)
print("Net Salary    :", net_salary)





```

# Student Performance Management System

```python
# Grade Criteria (Tuple)
grade_criteria = (
    "A : 90% and above",
    "B : 75-89%",
    "C : 60-74%",
    "D : 40-59%",
    "F : Below 40%"
)

students = {}

# Function to calculate total
def calculate_total(marks):
    return sum(marks)

# Function to calculate percentage
def calculate_percentage(total):
    return total / 5

# Function to assign grade
def assign_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 75:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 40:
        return "D"
    else:
        return "F"

# Function to display student details
def display_student(roll, student):
    print("----------------------------------")
    print("Roll No      :", roll)
    print("Name         :", student["Name"])
    print("Marks        :", student["Marks"])
    print("Total        :", student["Total"])
    print("Percentage   :", round(student["Percentage"], 2), "%")
    print("Grade        :", student["Grade"])

# Safe input for number of students
while True:
    try:
        n = int(input("\nEnter Number of Students: "))
        if n > 0:
            break
        else:
            print("Please enter a number greater than 0.")
    except ValueError:
        print("Invalid input! Please enter only numbers.")

# Input student details
for i in range(n):

    print(f"\nEnter Details of Student {i+1}")

    roll = input("Roll Number: ")
    name = input("Student Name: ")

    marks = []

    for j in range(5):
        while True:
            try:
                mark = float(input(f"Subject {j+1} Marks: "))
                if 0 <= mark <= 100:
                    marks.append(mark)
                    break
                else:
                    print("Marks should be between 0 and 100.")
            except ValueError:
                print("Please enter valid marks.")

    total = calculate_total(marks)
    percentage = calculate_percentage(total)
    grade = assign_grade(percentage)

    students[roll] = {
        "Name": name,
        "Marks": marks,
        "Total": total,
        "Percentage": percentage,
        "Grade": grade
    }

# Display all records
print("\n ALL STUDENT RECORDS ")

for roll, student in students.items():
    display_student(roll, student)

# Find topper
topper_roll = None
highest_percentage = -1

for roll, student in students.items():
    if student["Percentage"] > highest_percentage:
        highest_percentage = student["Percentage"]
        topper_roll = roll

# Display topper
if topper_roll is not None:
    print("\nTOPPER")
    display_student(topper_roll, students[topper_roll])

# Display grade criteria
print("\nGrade Criteria")
for item in grade_criteria:
    print(item)
```
