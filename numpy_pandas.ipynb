```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Mouse", "Keyboard", "Monitor"],
    "Price": [60000, 25000, 800, 1500, 12000]
}

df = pd.DataFrame(data)

df.to_csv("products.csv", index=False)

```

# Calculate average price using NumPy

```python

average_price = np.mean(df["Price"])
```

# Function to display products above average

```python

def above_average(df, avg):
    print("\nProducts Above Average Price:\n")

    for i in range(len(df)):
        if df.loc[i, "Price"] > avg:
            print(df.loc[i, "Product"], "-", df.loc[i, "Price"])

above_average(df, average_price)
```

Q2. Temperature Analysis using Pandas
Step 1: Create CSV File

```python
import pandas as pd

data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"],
    "Temperature": [18, 25, 35, 28, 15]
}

df = pd.DataFrame(data)

df.to_csv("temperature.csv", index=False)

print("temperature.csv created successfully!")
```

# Importing dataset and library

```python
import pandas as pd
import numpy as np
```

```python
# Read CSV
df = pd.read_csv("temperature.csv")
```

```python
print(df)
```

```python
# NumPy calculations
highest = np.max(df["Temperature"])
lowest = np.min(df["Temperature"])
average = np.mean(df["Temperature"])
```

```python
print("\nHighest Temperature =", highest)
print("Lowest Temperature =", lowest)
print("Average Temperature =", average)
```

```python
# Function to classify temperature
def classify(temp):
    if temp < 20:
        return "Cold"
    elif temp <= 30:
        return "Moderate"
    else:
        return "Hot"

print("\nTemperature Category")

for i in range(len(df)):
    day = df.loc[i, "Day"]
    temp = df.loc[i, "Temperature"]
    print(day, "-", temp, "°C -", classify(temp))
```
